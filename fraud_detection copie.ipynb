#### lat, long : latitude / longitude du client / carte bancaire (lieu d‚Äôorigine de la transaction)
#### merch_lat, merch_long latitude / longitude du marchand (o√π la transaction a eu lieu)



```python
# Distance entre client et marchand

def haversine_distance(lat1, lon1, lat2, lon2):
    # Rayon moyen de la Terre (km)
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

df["distance_km"] = haversine_distance(
    df["lat"], df["long"], df["merch_lat"], df["merch_long"]
)

df = df.drop(columns=["lat", "long", "merch_lat", "merch_long"])

display(df.head())
```

### üß† Preprocessing

```python
numeric_features = [
    "amt",
    "city_pop",
    "distance_km",
    "age",
    "year",
    "month",
    "day_of_week",
    "hour",
    "is_weekend"
]

categorical_features = [
    "merchant",
    "category",
    "gender",
    "state",
    "job"
]

drop_cols = [
    "cc_num",
    "first",
    "last",
    "street",
    "city",
    "zip",
    "trans_num",
    "unix_time"  # car redondant avec features d√©riv√©es
]

df_model = df.drop(columns=drop_cols)
X = df_model.drop(columns=[TARGET])
y = df_model[TARGET]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=RS, stratify = y)


numeric_transformer = Pipeline(
  steps=[
    ("scaler_num", StandardScaler()),
  ]
)

categorical_transformer = Pipeline(
  steps=[
    ("encoder_cat", OneHotEncoder(drop="first", handle_unknown="ignore")),#
  ]
)

preprocessor = ColumnTransformer(
  transformers=[
    ("num", numeric_transformer,     numeric_features),
    ("cat", categorical_transformer, categorical_features),
  ]
)

X_train = preprocessor.fit_transform(X_train)
X_test  = preprocessor.transform(X_test)
```

### üöÄ Premier mod√®le : Logistique regression notre "baseline"

```python

# max_iter=1000 nombre maximum d‚Äôit√©rations autoris√©es pour que l‚Äôalgorithme de r√©gression logistique converge vers un minimum de la fonction de co√ªt.

log_reg = LogisticRegression(max_iter=1000, class_weight="balanced")

log_reg.fit(X_train, y_train)

# Pr√©dictions
y_train_pred = log_reg.predict(X_train)
y_test_pred = log_reg.predict(X_test)

# Scores
print(f"train ---> f1-score : {f1_score(y_train, y_train_pred):.4f} \t precision_score : {precision_score(y_train, y_train_pred):.4f} \t recall_score : {recall_score(y_train, y_train_pred):.4f}")
print(f"test ----> f1-score : {f1_score(y_test, y_test_pred):.4f} \t precision_score : {precision_score(y_test, y_test_pred):.4f} \t recall_score : {recall_score(y_test, y_test_pred):.4f}")

# √âvaluation

# Matrice de confusion sur le train
print("Confusion matrix (train) :")
print(confusion_matrix(y_train, y_train_pred))

# Matrice de confusion sur le test
print("\nConfusion matrix (test) :")
print(confusion_matrix(y_test, y_test_pred))
```

```python
y_train_proba = log_reg.predict_proba(X_train)[:, 1]
thresholds = np.arange(0.01, 1.0, 0.01)
f1_scores = [f1_score(y_train, y_train_proba >= t) for t in thresholds]

best_threshold = thresholds[np.argmax(f1_scores)]
print(f"Meilleur seuil (threshold) : {best_threshold:.2f}")
```

```python
y_train_proba = log_reg.predict_proba(X_train)[:, 1]

y_train_pred_custom = (y_train_proba >= 0.98).astype(int)

f1_train = f1_score(y_train, y_train_pred_custom)
print(f"f1-score train avec seuil 0.98 : {f1_train:.4f}")
```

```python
# Probabilit√©s sur le test
y_test_proba = log_reg.predict_proba(X_test)[:, 1]


y_test_pred_custom = (y_test_proba >= 0.98).astype(int)

# F1-score
f1_test = f1_score(y_test, y_test_pred_custom)
print(f"f1-score test avec seuil 0.98 : {f1_test:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight="balanced_subsample",
    max_depth=10,             # limite la profondeur des arbres
    min_samples_split=50,     # min. d‚Äô√©chantillons pour scinder un noeud
    min_samples_leaf=20,      # min. d‚Äô√©chantillons par feuille
    n_jobs=-1                 # pour utiliser tous les c≈ìurs
)

rf.fit(X_train, y_train)

# Pr√©dictions
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)

# Scores
print(f"train ---> f1-score : {f1_score(y_train, y_train_pred):.4f} \t precision_score : {precision_score(y_train, y_train_pred):.4f} \t recall_score : {recall_score(y_train, y_train_pred):.4f}")
print(f"test ----> f1-score : {f1_score(y_test, y_test_pred):.4f} \t precision_score : {precision_score(y_test, y_test_pred):.4f} \t recall_score : {recall_score(y_test, y_test_pred):.4f}")
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=400,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=(len(y_train) - y_train.sum()) / y_train.sum(),  # ratio n√©gatifs/positifs
    eval_metric="logloss",
    n_jobs=-1,
    random_state=42
)

xgb.fit(X_train, y_train)

from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix

# Pr√©dictions brutes (classe 0/1)
y_train_pred = xgb.predict(X_train)
y_test_pred  = xgb.predict(X_test)

# Scores F1, pr√©cision, rappel
print(f"train ---> f1-score : {f1_score(y_train, y_train_pred):.4f} \t precision : {precision_score(y_train, y_train_pred):.4f} \t recall : {recall_score(y_train, y_train_pred):.4f}")
print(f"test ----> f1-score : {f1_score(y_test, y_test_pred):.4f} \t precision : {precision_score(y_test, y_test_pred):.4f} \t recall : {recall_score(y_test, y_test_pred):.4f}")

# Matrices de confusion (facultatif)
print("\nConfusion matrix (train):")
print(confusion_matrix(y_train, y_train_pred))

print("\nConfusion matrix (test):")
print(confusion_matrix(y_test, y_test_pred))
```

```python
import numpy as np
from sklearn.metrics import f1_score, precision_score, recall_score, confusion_matrix

# 1Ô∏è‚É£ Probabilit√©s de classe 1
y_proba_train = xgb.predict_proba(X_train)[:, 1]
y_proba_test  = xgb.predict_proba(X_test)[:, 1]

# 2Ô∏è‚É£ Recherche du meilleur seuil sur les PROBAS du test
best_f1, best_thresh = 0, 0
for thresh in np.arange(0.0, 1.0, 0.01):
    preds = (y_proba_test >= thresh).astype(int)   # ‚úÖ ici on utilise y_proba_test
    f1 = f1_score(y_test, preds)
    if f1 > best_f1:
        best_f1, best_thresh = f1, thresh

print(f"\nüîé Meilleur seuil : {best_thresh:.2f} avec F1 = {best_f1:.4f}")

# 3Ô∏è‚É£ Appliquer ce seuil aux proba train + test
y_train_pred_opt = (y_proba_train >= best_thresh).astype(int)
y_test_pred_opt  = (y_proba_test >= best_thresh).astype(int)

print(f"train ---> f1: {f1_score(y_train, y_train_pred_opt):.4f} "
      f"\t precision: {precision_score(y_train, y_train_pred_opt):.4f} "
      f"\t recall: {recall_score(y_train, y_train_pred_opt):.4f}")

print(f"test ----> f1: {f1_score(y_test, y_test_pred_opt):.4f} "
      f"\t precision: {precision_score(y_test, y_test_pred_opt):.4f} "
      f"\t recall: {recall_score(y_test, y_test_pred_opt):.4f}")

print("\nConfusion matrix (test):")
print(confusion_matrix(y_test, y_test_pred_opt))
```

```python

# Ratio n√©gatifs / positifs pour scale_pos_weight
scale_pos_weight = (len(y_train) - y_train.sum()) / y_train.sum()

xgb = XGBClassifier(
    n_estimators=400,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    eval_metric="logloss",
    n_jobs=-1,
    random_state=42
)

from sklearn.pipeline import Pipeline

model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", xgb),
])
```

```python
model.fit(X_train, y_train)
```

```python
# Pr√©dictions classes
y_train_pred = model.predict(X_train)
y_test_pred  = model.predict(X_test)

print(f"train ---> f1-score : {f1_score(y_train, y_train_pred):.4f} \t precision : {precision_score(y_train, y_train_pred):.4f} \t recall : {recall_score(y_train, y_train_pred):.4f}")
print(f"test ----> f1-score : {f1_score(y_test, y_test_pred):.4f} \t precision : {precision_score(y_test, y_test_pred):.4f} \t recall : {recall_score(y_test, y_test_pred):.4f}")
```
